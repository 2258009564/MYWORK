```python
import pandas as pd
import openpyxl
import math
```

```python
def triangle_area(a_side, b_side, c_side):
    p = (a_side + b_side + c_side) / 2
    area = math.sqrt(p * (p - a_side) * (p - b_side) * (p - c_side))
    return area
```

```python
excel_path = '三角面积(1).xlsx'
df1 = pd.read_excel(excel_path )
df1
```

```python
list1 = []
for i in range(len(df1)):
    a , b , c = int(df1['木棒1'][i]), int(df1['木棒2'][i]), int(df1['木棒3'][i])
    if a + b > c and a + c > b and b + c > a:
        list1.append(triangle_area(a, b, c))
    else:
        list1.append(1000)
list1    
```

```python
min_area = min(list1)
min_index = list1.index(min_area) + 1
```
